```python
import pandas as pd
import numpy as np

# Load your cleaned CSV
df = pd.read_csv(r"C:\Users\hezek\Documents\gas_2025_clean.csv")

# Remove TOTAL row
df = df[~df['MONTHS'].str.contains("TOTAL", na=False)]

# Numeric columns to clean
num_cols = [
    'AG PRODUCTION\n(MMSCF)',
    'NAG PRODUCTION\n(MMSCF)',
    'TOTAL GAS\nPRODUCTION\n(MMSCF)',
    'FIELD USE\n(MMSCF)',
    'DOMESTIC SALES\n(MMSCF)',
    'EXPORT SALES\n(MMSCF)',
    'TOTAL GAS UTILISED\n(MMSCF)',
    'TOTAL GAS\nFLARED (MMSCF)',
    'GAS SHRINKAGE'
]

# Clean numbers (remove commas, convert dash to 0)
for c in num_cols:
    df[c] = (
        df[c]
        .astype(str)
        .str.replace(",", "")   # remove commas
        .str.replace("-", "0")  # replace dash
        .str.strip()
    )
    df[c] = pd.to_numeric(df[c], errors="coerce")

# Clean percentages
df['% UTILIZED'] = df['% UTILIZED'].str.replace("%", "").astype(float)
df['% FLARED'] = df['% FLARED'].str.replace("%", "").astype(float)

# Fix MONTHS column → datetime
df['MONTHS'] = df['MONTHS'].str.title()   # "JANUARY" → "January"
df['month_dt'] = pd.to_datetime(df['MONTHS'], format="%B")
df['month_dt'] = df['month_dt'].apply(lambda d: d.replace(year=2025))

df = df.sort_values("month_dt")
df.head()

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

plt.plot(df['month_dt'], df['TOTAL GAS\nPRODUCTION\n(MMSCF)'],
         marker='o', label='Total Production')

plt.plot(df['month_dt'], df['AG PRODUCTION\n(MMSCF)'],
         marker='o', linestyle='--', label='AG Production')

plt.plot(df['month_dt'], df['NAG PRODUCTION\n(MMSCF)'],
         marker='o', linestyle=':', label='NAG Production')

plt.title("Monthly Gas Production Trend (2025)")
plt.xlabel("Month")
plt.ylabel("Volume (MMSCF)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
mean_util = df['% UTILIZED'].astype(float).mean()

best_month = df.loc[df['% UTILIZED'].idxmax(), 'MONTHS']
worst_month = df.loc[df['% UTILIZED'].idxmin(), 'MONTHS']

print("Mean Utilisation:", mean_util)
print("Best Month:", best_month)
print("Worst Month:", worst_month)

```

```python
# --- STEP 1: RENAME COLUMNS ---
# Rename the complex, multi-line column names to simple, easy-to-use variables
df = df.rename(columns={
    'TOTAL GAS\nPRODUCTION\n(MMSCF)': 'total_mmscf',
    'TOTAL GAS UTILISED\n(MMSCF)': 'utilised_mmscf',
    'TOTAL GAS\nFLARED (MMSCF)': 'flared_mmscf'
})

# --- STEP 2: CUMULATIVE CALCULATION ---
# Calculate the running total for each metric
df['cum_prod'] = df['total_mmscf'].cumsum()
df['cum_util'] = df['utilised_mmscf'].cumsum()
df['cum_flare'] = df['flared_mmscf'].cumsum()

# --- STEP 3: PLOTTING ---
plt.figure(figsize=(10,5))
plt.plot(df['month_dt'], df['cum_prod'], marker='o', label='Cumulative Production')
plt.plot(df['month_dt'], df['cum_util'], marker='o', label='Cumulative Utilisation')
plt.plot(df['month_dt'], df['cum_flare'], marker='o', label='Cumulative Flaring')

plt.title("Cumulative Gas Metrics (2025)")
plt.xlabel("Month")
plt.ylabel("MMSCF")
plt.grid(True)
plt.legend()
plt.show()
```
